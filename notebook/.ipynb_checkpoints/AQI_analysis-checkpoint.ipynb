# AQI / Pollution Analysis
**Purpose:** Load cleaned pollution file, compute simple PM2.5-derived AQI, plot national trend and top polluted cities. Contains placeholder for full AQI computation.


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

base = r"C:\Users\Vishakha Khokhar\Desktop\Project"
clean_path = os.path.join(base, "cleaned_data")
visuals = os.path.join(base, "visuals")
os.makedirs(visuals, exist_ok=True)
```

```python
aqi_file = os.path.join(clean_path, "aqi_india_clean.csv")
print("Loading:", aqi_file)
poll = pd.read_csv(aqi_file)
poll.head()
```

```python
# Standardize column names we used earlier: Pollutant, Value, city, state, station, Date (if present)
print("Columns:", poll.columns.tolist())
poll.rename(columns={c:c.strip() for c in poll.columns}, inplace=True)

# Ensure numeric value
poll['Value'] = pd.to_numeric(poll['Value'], errors='coerce')
poll = poll.dropna(subset=['Value'])
```

```python
# Simple PM2.5->AQI mapping (illustrative). Replace with official breakpoints for production.
def pm25_to_aqi(x):
    if x <= 30: return 50
    elif x <= 60: return 100
    elif x <= 90: return 200
    elif x <= 120: return 300
    elif x <= 250: return 400
    else: return 500

pm25 = poll[poll['Pollutant'].str.upper() == 'PM2.5'.upper()].copy()
if pm25.empty:
    print("No PM2.5 rows found in file. You may need to compute AQI from other pollutants or check pollutant labels.")
else:
    # if Date exists, derive Year
    if 'Date' in pm25.columns:
        pm25['Year'] = pd.to_datetime(pm25['Date'], errors='coerce').dt.year
    # compute AQI
    pm25['AQI'] = pm25['Value'].apply(pm25_to_aqi)
    # national average (by year)
    if 'Year' in pm25.columns:
        yearly_aqi = pm25.groupby('Year')['AQI'].mean().reset_index()
        plt.figure(figsize=(10,5))
        plt.plot(yearly_aqi['Year'], yearly_aqi['AQI'], marker='o')
        plt.title("India Average AQI (derived from PM2.5)")
        plt.xlabel("Year"); plt.ylabel("AQI")
        plt.grid(alpha=0.3)
        fn = os.path.join(visuals, "aqi_trend_pm25_from_notebook.png")
        plt.savefig(fn, bbox_inches='tight'); plt.show()
        print("Saved:", fn)
```

```python
if not pm25.empty:
    top_cities = (pm25.groupby('city')['Value'].mean()
                       .sort_values(ascending=False).head(10).reset_index())
    plt.figure(figsize=(8,6))
    plt.barh(top_cities['city'][::-1], top_cities['Value'][::-1])
    plt.xlabel("PM2.5 (avg)"); plt.title("Top 10 Cities by PM2.5")
    fn = os.path.join(visuals, "aqi_city_top10_from_notebook.png")
    plt.savefig(fn, bbox_inches='tight'); plt.show()
    print("Saved:", fn)
```
