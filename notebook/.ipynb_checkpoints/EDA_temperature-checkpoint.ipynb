**Purpose:** Load the combined temperature grid file, inspect the data, compute yearly global mean temperature, plot trends and sample spatial distribution.



```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

base = r"C:\Users\Vishakha Khokhar\Desktop\Project"
clean_path = os.path.join(base, "cleaned_data")
visuals = os.path.join(base, "visuals")
os.makedirs(visuals, exist_ok=True)
```

```python
temp_file = os.path.join(clean_path, "temperature_global_clean.csv")
print("Loading:", temp_file)
temp = pd.read_csv(temp_file)
temp.head()
```

```python
temp_file = os.path.join(clean_path, "temperature_global_clean.csv")
print("Loading:", temp_file)
temp = pd.read_csv(temp_file)
temp.head()
```

```python
print("Shape:", temp.shape)
print("\nColumns:", temp.columns.tolist())
print("\nDtypes:\n", temp.dtypes)
print("\nMissing values per column:\n", temp.isna().sum())
```

```python
temp['Temperature'] = pd.to_numeric(temp['Temperature'], errors='coerce')
temp['Latitude'] = pd.to_numeric(temp['Latitude'], errors='coerce')
temp['Longitude'] = pd.to_numeric(temp['Longitude'], errors='coerce')
temp['Year'] = pd.to_numeric(temp['Year'], errors='coerce')
print("After convert, missing temps:", temp['Temperature'].isna().sum())
temp = temp.dropna(subset=['Temperature','Year','Latitude','Longitude'])
```

```python
yearly_temp = temp.groupby('Year')['Temperature'].mean().reset_index()
yearly_temp.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(yearly_temp['Year'], yearly_temp['Temperature'], marker='o')
plt.title("Global Yearly Mean Temperature")
plt.xlabel("Year"); plt.ylabel("Mean Temperature")
plt.grid(alpha=0.3)
fn = os.path.join(visuals, "temp_yearly_mean_from_notebook.png")
plt.savefig(fn, bbox_inches='tight')
plt.show()
print("Saved:", fn)
```

**Observations:**  
- Check the plot for long-term trends and acceleration periods.  
- Note any sudden jumps or outliers; investigate if they are data issues or real events.


```python
# pick a sample year (median)
years = sorted(temp['Year'].unique())
sample_year = years[len(years)//2]
s = temp[temp['Year'] == sample_year]

plt.figure(figsize=(12,6))
sc = plt.scatter(s['Longitude'], s['Latitude'], c=s['Temperature'], s=6)
plt.colorbar(sc, label='Temperature')
plt.title(f"Temperature grid scatter ({sample_year})")
plt.xlabel("Longitude"); plt.ylabel("Latitude")
fn = os.path.join(visuals, f"temp_world_scatter_{sample_year}_notebook.png")
plt.savefig(fn, bbox_inches='tight')
plt.show()
print("Saved:", fn)
```
